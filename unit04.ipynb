## Visualizing a function

### Run the imports before the other code cells

```python
# define shorthands for imports
import matplotlib.pyplot as plt 
import numpy as np
```

### sin and cos functions

```python
x = np.arange(0, 6, 0.1) # array [0, 0.1, ..., 5.9]

plt.plot(x, np.sin(x), label="sin") # solid line for sin
plt.plot(x, np.cos(x), linestyle="--", label="cos") # dashed line for cos

plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show() # display the figure
```

### floor and ceiling functions

```python
# create an evenly spaced array of size 500 for interval [0, 5]
x = np.linspace(0, 5, 500)

plt.plot(x, np.floor(x), '-') # solid line for floor
plt.plot(x, np.ceil(x), '--') # dashed line for ceiling

plt.show() # display the figure
```

```python
print("floor(2.3) =", np.floor(2.3))
print("floor(-2.3) =", np.floor(-2.3))
print("ceil(2.3) =", np.ceil(2.3))
print("ceil(-2.3) =", np.ceil(-2.3))
```

### Exponential and logarithmic functions

```python
x1 = np.linspace(-8, 2, 100)
x2 = np.linspace(0.01, 8, 100)
x3 = np.linspace(-5, 8, 100)

plt.plot(x1, np.exp(x1), '-', label='y=exp(x)')
plt.plot(x2, np.log(x2), '-.', label='y=log(x)')
plt.plot(x3, x3, '--', label='y=x')

plt.legend(loc='upper right')
plt.axis('equal')
plt.grid(True)
plt.show()
```

### Modulo function

```python
x = np.linspace(-10, 10, 500)
plt.plot(x, np.mod(x,3))
plt.show()
```
