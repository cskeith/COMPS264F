## Generating permutations

```
from itertools import permutations 

s = ['John', 'Mary', 'Peter', 'Tom']

# Generate all permutations
perm = permutations(s) 
for i in perm:
    print(i)
```

```
# Generate all 3-permutations
perm = permutations(s, 3) 
for i in perm:
    print(i)
```

## Number of permutations

```
from math import factorial

def nPr(n, r):
    return factorial(n) // factorial(n-r)  # Use // for integer division (floor division)

print(nPr(8,3))
```

# Generating combinations

```
from itertools import combinations 

# Generate all 2-combinations
comb = combinations([1, 2, 3], 2) 
for i in comb:
    print(i)
```

# Number of combinations

```
from scipy.special import comb

print(comb(10, 3, exact=True))
```

# Combinations with repetition

```
from itertools import combinations_with_replacement 
  
# Generate all 2-combinations with repeated elements
comb = combinations_with_replacement([1, 2, 3], 2) 
for i in comb: 
    print(i) 
```

```
from scipy.special import comb

print(comb(4+4-1, 4, exact=True))
print(comb(4, 4, exact=True, repetition=True))
```
